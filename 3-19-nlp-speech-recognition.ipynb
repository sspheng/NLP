```python
#pip install SpeechRecognition
#pip install PyAudio
#import library
import speech_recognition as sr
```

```python
#Initiаlize  reсоgnizer  сlаss  (fоr  reсоgnizing  the  sрeeсh)
r = sr.Recognizer()
```

```python
# Reading Audio file as source
# listening  the  аudiо  file  аnd  stоre  in  аudiо_text  vаriаble

with sr.AudioFile('./audiofiles/harvard.wav') as source:
    audio_data = r.listen(source)
    
# recoginize_() method will throw a request error if the API is unreachable, hence using exception handling
    try:  
        # using google speech recognition
        text = r.recognize_google(audio_data)
        
        print('Converting audio transcripts into text ...')
        print(text)
        
    except:
         print('Sorry.. run again...Check Internect Connection')
```

```python
with sr.Microphone() as source:
    
    print("Speak Now...\n")
    
    # read the audio data from the default microphone
    audio_data = r.record(source, duration=5)
    
    print("Recognizing...\n")
    
    # convert speech to text
    text = r.recognize_google(audio_data)
    
    print(text)
```

```python
print(audio_data.frame_data)
```

```python
# write audio data to a WAV file
with open("./audiofiles/microphone-results.wav", "wb") as f:
    f.write(audio_data.get_wav_data())
```

```python
# Reading Audio file as source
# listening  the  аudiо  file  аnd  stоre  in  аudiо_text  vаriаble

with sr.AudioFile('./audiofiles/microphone-results.wav') as source:
    audio_data = r.listen(source)
    
# recoginize_() method will throw a request error if the API is unreachable, hence using exception handling
    try:
        
        # using google speech recognition
        text = r.recognize_google(audio_data, language="es-ES")
        
        print('Converting audio transcripts into text ...')
        print(text)
        
    except:
         print('Sorry.. run again...')
```
